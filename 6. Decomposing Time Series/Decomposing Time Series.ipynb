# Decomposing Time Series

#### Additive Model
* y(t) = Level + Trend + Seasonality + Noise

#### Multiplicative Model
* y(t) = Level * Trend + Seasonality + Noise

```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
miles_decompose_df = pd.read_csv('E:/Time Series Analysis and Forecasting Using Python/1. Time Series - Basics(Data and Code)/us-airlines-monthly-aircraft-miles-flown.csv', header = 0, parse_dates = [0])
```

```python
miles_decompose_df.head()
```

```python
miles_decompose_df.index = miles_decompose_df.Month
```

```python
result = seasonal_decompose(miles_decompose_df.MilesMM, model = 'additive')
```

```python
result.plot()
```

```python
result2 = seasonal_decompose(miles_decompose_df.MilesMM, model = 'multiplicative')
```

```python
result2.plot()
```
