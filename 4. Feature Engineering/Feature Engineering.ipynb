# Feature Engineering

* Date Time Features
* Lag Features
* Window Features
* Expanding Feature

```python
import pandas as pd
```

```python
df =  pd.read_csv('E:/Time Series Analysis and Forecasting Using Python/1. Time Series - Basics(Data and Code)/daily-total-female-births-CA.csv', header = 0, parse_dates = [0])
```

```python
df.head()
```

```python
features = df.copy()
```

```python
features.head()
```

## Date Time Features

```python
features['Year'] = df.date.dt.year
```

```python
features['Month'] = df.date.dt.month
```

```python
features['Day'] = df.date.dt.day
```

```python
features.head()
```

## Lag features

* Delayed Values
* N-Previous days values 

```python
features['Lag1'] = df.births.shift(1)
```

```python
#To get the value on same day but Last Year so we pass 365 as input as a year contains 365 days.
features['Lag2'] = df.births.shift(365)
```

```python
features.head()
```

## Window Features

* We provide the Window size to Consider the Particular number of Time Period

```python
features["Roll_mean"] = df.births.rolling(window=2).mean() #Window Size is the Number of Periods we want. here i have taken Previous 2 days values.
```

```python
features.head()
```

```python
# To get Maximum values of Previous 3 Period
features['Roll_max'] = df.births.rolling(window=3).max()
```

```python
features.head()
```

## Expanding Features

* Considers the Whole Dataset

```python
features['Expand_Max'] = df.births.expanding().max()
```
