# Differencing

* Removing Trend and Seasonality
* For removing Trend, we give periods = 1
* for Removing Seasonality, we give periods = 12(For Monthly data)

```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
from statsmodels.tsa.seasonal import seasonal_decompose 
```

```python
miles_df = pd.read_csv('E:/Time Series Analysis and Forecasting Using Python/1. Time Series - Basics(Data and Code)/us-airlines-monthly-aircraft-miles-flown.csv', header = 0, parse_dates = [0])
```

```python
miles_df.head()
```

```python
miles_df['Lag1'] = miles_df.MilesMM.shift(1)
```

```python
miles_df.head()
```

```python
miles_df['Miles_diff_1']=miles_df.MilesMM.diff(periods=1)
```

```python
miles_df.head()
```

```python
miles_df.index = miles_df.Month
result_a = seasonal_decompose(miles_df.MilesMM, model='additive')
result_a.plot()
```

```python
result_b = seasonal_decompose(miles_df.iloc[1:,3], model='additive')
result_b.plot()
```

```python
miles_df.MilesMM.plot()
```

```python
miles_df.Lag1.plot()
```

```python
miles_df.Miles_diff_1.plot()
```

```python
miles_df['miles_diff_12']= miles_df.MilesMM.diff(periods=12)
```

```python
miles_df.head(15)
```

```python
miles_df.miles_diff_12.plot()
```

```python
result_c = seasonal_decompose(miles_df.iloc[12:,5], model='additive')
result_c.plot()
```
