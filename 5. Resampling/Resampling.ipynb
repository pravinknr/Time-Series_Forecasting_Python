# Resampling

* Downsampling - Combining the Observations to a Meaningful Data.
* Upsampling - Expanding the Observations to a Meaningful Data.

```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
miles_df = pd.read_csv('E:/Time Series Analysis and Forecasting Using Python/1. Time Series - Basics(Data and Code)/us-airlines-monthly-aircraft-miles-flown.csv', header = 0, parse_dates = [0])
```

```python
miles_df.head()
```

## Downsampling

```python
# Downsampling into Quarterly Data. 'Q' is an built-in Keyword for 'Quarter'
quarter_df = miles_df.resample('Q', on='Month').mean()
```

```python
quarter_df.head()
```

```python
# Downsampling into Yearly Data. 'A' is an Built-in Keyword for "Annual" or 'Year'.
yearly_df = miles_df.resample('A', on='Month').sum()
```

```python
yearly_df.head()
```

## Upsampling

```python
# 'D'is for Daily Data
upsampled_df = miles_df.resample('D', on = 'Month').mean()
```

```python
upsampled_df.head(35)
```

```python
# To Fill the Empty cells linearly we use Interploation method with linear method
interpolated_miles_df = upsampled_df.interpolate(method='linear')
```

```python
interpolated_miles_df.head(35)
```

```python
interpolated_miles_df.plot()
```

```python
# Polynomial method Interpolation
poly_interpolated_miles_df = upsampled_df.interpolate(method='spline', order=2)
```

```python
poly_interpolated_miles_df.head()
```

```python
poly_interpolated_miles_df.plot()
```
